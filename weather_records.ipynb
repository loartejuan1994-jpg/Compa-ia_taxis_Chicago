```python
# Paso 1: - Paso 1: Extraer datos del clima desde sitio web

import pandas as pd                # Maneja datos en tablas (DataFrames)
import requests                    # Conecta al sitio web y descarga su contenido
from bs4 import BeautifulSoup      # Analiza y navega el HTML descargado

URL = 'https://practicum-content.s3.us-west-1.amazonaws.com/data-analyst-eng/moved_chicago_weather_2017.html'
req = requests.get(URL)                                # Hace la petición HTTP y descarga la página
soup = BeautifulSoup(req.content, 'html.parser')       # Convierte el HTML en objeto navegable 


# Busca la tabla específica por su atributo id
table = soup.find('table', attrs={"id": "weather_records"})  # más preciso que solo find('table')

# Extrae encabezados de la tabla
heading_table =[]                                      # Lista vacía para guardar los encabezados
for row in table.find_all ('th'):                      # Recorre todas las etiquetas <th> (encabezados)
    heading_table.append(row.text)                     # Extrae el texto de cada encabezado y lo agrega a la lista
    
# Extraer datos de la tabla
content = []                                           # Lista vacía para guardar las filas de datos
for row in table.find_all('tr'):                       # Recorre toda las etiquetas <tr> (filas)
    if row.find_all('td'):                             # Verifica que la fila tenga celdas <td> (ignora encabezados)
        content.append([cell.text for cell in row.find_all('td')])  # Agrega cada fila como lista

# Crear DataFrame con los datos extraídos anteriormente
weather_records = pd.DataFrame(content, columns=heading_table)  # Usa encabezados como nombres de columnas

print(weather_records)
```

```python
# Estructura de datos

print('Información General del DataFrame')
print(weather_records.info())
print('--'* 25)

print('Primeras 5 filas')
print(weather_records.head())
print('--' * 25)

print('Valores nulos por columnas')
print(weather_records.isnull().sum())
print('--' * 25)

print(F'Número total de filas: {len(weather_records)}')
print(f'\nFilas duplicadas: {weather_records.duplicated().sum()}')
print('--' * 25)

print('Tipos de datos')
print(weather_records.dtypes)
```

```python
# Convertir tipos de datos

weather_records['Date and time'] = pd.to_datetime(weather_records['Date and time'])

weather_records['Temperature'] = pd.to_numeric(weather_records['Temperature'], errors='coerce')

print("Tipo de datos despues de convertirlos:")
print(weather_records.dtypes)
```
